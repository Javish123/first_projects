**Se debe hacer Scroll Down para que la página Principal de Búsqueda cague más links:**

```python
#chromedriver = "/Users/monita/ds/metis/metisgh/bootcamp/P4_ArtRecommender/chromedriver"
chromedriver = "../utils/chromedriver"
os.environ["webdriver.chrome.driver"] = chromedriver
driver = webdriver.Chrome(chromedriver)
```

```python
#Intento con Scroll y Body Height --> No funciona para esta página
SCROLL_PAUSE_TIME = 1
driver.get("http://www.museodelprado.es/coleccion/obras-de-arte")

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")
time.sleep(5)
while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
#Intento por teclado: Keys.PAGE_DOWN
loop_counter = 0
driver.get("http://www.museodelprado.es/coleccion/obras-de-arte")
time.sleep(5)
while loop_counter < 1000:
#     driver.execute_script("window.scrollTo(0, 300)")
#     driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    scroll = driver.find_element_by_tag_name('body')
    scroll.send_keys(Keys.PAGE_DOWN)
    time.sleep(2)
    if loop_counter%100 == 0:
        print(loop_counter)
    loop_counter+=1
```

```python
url_list = []
#driver.get("http://www.museodelprado.es/coleccion/obras-de-arte")
soup = BeautifulSoup(driver.page_source,"html5lib")
lista_ref = soup.findAll(class_="presentacion-mosaico")
for r in lista_ref :
    url_list.append(r.find("a").get('href'))
print(len(url_list))
url_list
```

```python
prado_list = []
for w in url_list[-5:] :
    try :
        response = requests.get(w)
        page = response.text
    except TimeoutException :
        print("Timed Out " + str(c) + ": " + w)
        response = requests.get(w)
        page = response.text
        continue
    soup = BeautifulSoup(page,"html5lib")
    title = soup.find(class_="obra").find("h1").text
    url = w
    desc = ''
    if soup.find(class_="obra").find(class_="colapsado normal") :
        desc = soup.find(class_="obra").find(class_="colapsado normal").text
    prado_list.append((title,url,desc))
    if (c%500 == 0) :
        print(c)
    c+=1
    time.sleep(.2)
```
